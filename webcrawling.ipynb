# 카카오맵에서 검색

```python
# 카카오 맵 접속
driver = webdriver.Chrome('C:\\chrome\\chromedriver')

driver.get('https://map.kakao.com/')
```

```python
#데이터 불러오기

train = pd.read_csv("C:\\Users\\acorn\\python\\city\\train_x_y_gu.csv", encoding = 'utf-8', index_col = 0)
```

```python
train['dong_apt'] = train['dong']+' '+train['apt']
```

```python
train.head(10)
```

```python
addr_name = list(train['dong_apt'].head(10))
```

```python
# 검색한 결과 저장하기
from tqdm import tqdm_notebook
import time

jibun_addr = []

try :
    for addr in tqdm_notebook(addr_name):
        # 키워드 입력
        elem = driver.find_element_by_id('search.keyword.query')
        elem.clear()
        elem.send_keys(addr)

        # 크롤링을 통해서 락이 걸리지 않도록 하기 위함
        time.sleep(2)

        # 엔터 누르기
        xp = '//*[@id="search.keyword.submit"]'
        driver.find_element_by_xpath(xp).send_keys(Keys.ENTER)

        time.sleep(1)

        # 지번 주소 저장하기
        addr = driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[1]/div[5]/div[2]/p[2]')
        jibun_addr.append(addr.text)


        time.sleep(1)
except NoSuchElementException :
    print('검색 결과가 없습니다.')
    jibun_addr.append('None')

finally :
    driver.close()
```
